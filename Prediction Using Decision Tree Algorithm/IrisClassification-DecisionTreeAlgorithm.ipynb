### `Making X and y features`

```python
X = df.drop("Species", axis=1)
y = df["Species"] 
```

```python
X.head()
```

```python
y.head()
```

```python
X.dtypes
```

```python
y.dtypes
```

### `Note:` 

> `X` has all `numeric values` but `y` has `categorical value` which must be `encoded` into `numeric value i.e. machine-understandable value.`

For this, we use:
> `LabelEncoder()`

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Encoded species"] = le.fit_transform(y.ravel()) # Appending to our dataset
y = le.fit_transform(y.ravel())
y
```

The `ravel()` function is used to create a contiguous flattened array. It returns a 1-D array, containing the elements of the input.

```python
df.head()
```

From this, we can see encoding as:
    
`0` == `Iris-Setosa`

`1` == `Iris-Versicolor`

`2` == `Iris-Virginica`

```python
Encoded_class = pd.DataFrame({'Species': ["Iris-setosa", "Iris-versicolor", "Iris-virginica"],
                              'Encoded': [0, 1, 2]})
Encoded_class
```

# `STEP-6:` Preparing Training & Test Data and Fitting to the Model

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

np.random.seed(65)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2)
clf = DecisionTreeClassifier(criterion="gini", min_samples_split=2)
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
print(f"Model Accuracy on Test Data = {score*100:.2f}%")
```

# `STEP-7:` Model Evaluation

Metrics Used for Model Evalaution:
    
> a) `Classification Report`

> b) `Confusion Matrix`

> c) `Accuracy Score`

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_preds = clf.predict(X_test)
print(f"Classfifcation Report:\n\n{classification_report(y_test, y_preds)}")
```

```python
cf_matrix = confusion_matrix(y_test, y_preds)
print(f"Confusion Matrix:\n\n{cf_matrix}")
```

### `Visualizing Confusion Matrix`

```python
fig, ax = plt.subplots(figsize=(6,4))
sns.heatmap(cf_matrix, annot=True, cmap='coolwarm')
fig.suptitle(t="Confusion Matrix",
             color="orange",
             fontsize=16);
ax.set(xlabel="Predicted Label",
       ylabel="Actual Label");
```

```python
print(f"Accuracy Score:\n\n{accuracy_score(y_test, y_preds)*100:.2f}%")
```
